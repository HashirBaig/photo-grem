```python
import sys
import os
try:
    from osgeo import gdal
except:
    sys.exit("ERROR: cannot find one or more of the required module(s)")
```

```python
file_type = 'DSM'
```

```python
# Check valid file path
# input_file_path = f"./data/image/exported_image_{file_type}.tif"
input_file_path = f"./data/image/Reproj_AHN_{file_type}.tif"

if os.path.isfile(input_file_path):
    # If the file is valid, get and print the absolute path
    print("File is found. File path: " + os.path.abspath(input_file_path))
else:
    # Terminate the program if file is not valid
    raise FileNotFoundError(
        "File not found in this path :" + os.path.abspath(input_file_path))
```

```python
output_cog_path = f"./data/image/output_ahn_{file_type}_cog.tif"
# intermediate_cog_path = f"./data/image/temp_{file_type}_cog.tif"
```

```python
# Generate overviews (optional but recommended for COG)
def add_overviews(tif_path):
    ds = gdal.Open(tif_path, gdal.GA_Update)
    if ds:
        ds.BuildOverviews("NEAREST", [2, 4, 8, 16, 32])
        ds = None
    else:
        raise RuntimeError("Failed to open dataset for overviews.")

# Convert to Cloud Optimized GeoTIFF
def convert_to_cog(input_path, output_path):
    translate_options = gdal.TranslateOptions(
        format="COG",
        creationOptions=[
            "COMPRESS=LZW",            # LZW compression
            "BLOCKSIZE=512",           # Tile size
            "OVERVIEWS=IGNORE_EXISTING"  # Use existing or generate new
        ]
    )
    gdal.Translate(destName=output_path, srcDS=input_path, options=translate_options)
```

```python
print("📦 Preparing TIFF with JPEG compression and overviews...")
add_overviews(input_file_path)
```

```python
print("🚀 Creating optimized Cloud Optimized GeoTIFF...")
convert_to_cog(input_file_path, output_cog_path)

print("✅ Done. Output file:", output_cog_path)
```
